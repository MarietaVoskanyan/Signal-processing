```python
import numpy as np
import matplotlib.pyplot as plt

# Հիմնական պարամետրեր
F = 5       # Սինուսոիդի հաճախականություն, Hz
T = 1       # Տևողություն, վայրկյան
Fs_list = [20, 50, 100]  # Նմուշավորման հաճախականություններ, Hz

# Կատարել նմուշավորում և գրաֆիկներ
plt.figure(figsize=(12, 8))

for i, Fs in enumerate(Fs_list, 1):
    t = np.arange(0, T, 1/Fs)       # Նմուշավորված ժամանակ
    y = np.sin(2 * np.pi * F * t)   # Սինուսոիդալ ազդանշան
    
    plt.subplot(3, 1, i)
    plt.stem(t, y)
    plt.title(f'Sampled at Fs = {Fs} Hz')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Պարամետրեր
F = 15       # Սինուսոիդի հաճախականություն, Hz
T = 1        # Տևողություն, վայրկյան
Fs_list = [20, 40]  # Նմուշավորման հաճախականություններ

plt.figure(figsize=(10, 6))

for i, Fs in enumerate(Fs_list, 1):
    t = np.arange(0, T, 1/Fs)           # Նմուշավորված ժամանակ
    y = np.sin(2 * np.pi * F * t)       # Սինուսոիդ
    
    plt.subplot(2, 1, i)
    plt.stem(t, y)
    plt.title(f'Sampled at Fs = {Fs} Hz')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Պարամետրեր
F = 10        # Սինուսոիդի հաճախականություն, Hz
T = 1         # Տևողություն, վայրկյան
Fs = 50       # Նմուշավորման հաճախականություն, Hz

# Նմուշավորված ազդանշան
t_sampled = np.arange(0, T, 1/Fs)
y_sampled = np.sin(2 * np.pi * F * t_sampled)
t_fine = np.linspace(t_sampled[0], t_sampled[-1], 1000)

interpolator = interp1d(t_sampled, y_sampled, kind='cubic', fill_value="extrapolate")
t_fine = np.linspace(0, T, 1000)
y_restored = interpolator(t_fine)

# Գրաֆիկների ստեղծում
plt.figure(figsize=(10,6))
plt.plot(t_fine, y_restored, label='Restored signal (interpolated)', color='blue')
plt.stem(t_sampled, y_sampled, linefmt='r-', markerfmt='ro', basefmt='r-', label='Sampled signal')
plt.title('Original Sampled Signal and Restored Analog Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Պարամետրեր
Fs = 20        # Նմուշավորման հաճախականություն, Hz
T = 1          # Տևողություն, վայրկյան
t = np.arange(0, T, 1/Fs)  # Նմուշավորված ժամանակ

# Արդյունաբեր ազդանշան 2 բաղադրիչներով
y = np.sin(2*np.pi*3*t) + np.sin(2*np.pi*15*t)

# Գրաֆիկ՝ նմուշավորված ազդանշան
plt.figure(figsize=(10,4))
plt.stem(t, y)
plt.title('Sampled signal (3Hz + 15Hz) at Fs=20Hz')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Վերականգնում ինտերպոլյացիայով
t_fine = np.linspace(t[0], t[-1], 1000)
interpolator = interp1d(t, y, kind='cubic')
y_restored = interpolator(t_fine)

# Գրաֆիկ՝ վերականգնված ազդանշան
plt.figure(figsize=(10,4))
plt.plot(t_fine, y_restored, label='Restored signal (interpolated)')
plt.stem(t, y, linefmt='r-', markerfmt='ro', basefmt='r-', label='Sampled points')
plt.title('Interpolated Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
